### 결측치 처리
- null 이 있는 모든 행을 제거 df.dropna(inplace=True)
- null 이 있는 특정 열의 행만 제거 df.dropna(subset= [''])
- null 이 있는 모든 열을 제거 df.drop(['', ''], axis = 1)
- 결측값을 다른 값으로 채우기 fillna(df.median(), inplace = True)

```python
df_wine.fillna(df_wine.median(), inplace=True)
# df_wine.dropna(inplace=True)
# df_wine.dropna(subset=['alcohol'])
# df_wine.fillna(-99)
```

### 스케일링의 종류 (sklearn의 preprocessing을 이용)
1. 표준화 스케일링 - 평균이 0 표준편차가 1이 되게 변형
2. 로버스트 스케일링
3. 최소-최대 스케일링
4. 정규화 스케일링

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

### 3.1.2 Standard Scaling

$$
z = \frac{x - \mu}{\sigma}
$$



```python
st_scaler = StandardScaler()
st_scaler.fit(df_wine)
st_scaled = st_scaler.transform(df_wine) # transform: 스케일러가 fit을 통해 학습한 정보를 통해 데이터를 변화너, 즉 스케일링 실행
st_scaled # output: numpy array (no index, no column names)
```

```python
st_scaled = pd.DataFrame(st_scaled, columns = df_wine.columns)
st_scaled
```

```python
round(st_scaled.describe(), 2)
```

### 3.1.3 Robust Scaling
- 평균과 표준편차 대신 사분위값을 이용하여 계산
- 데이터의 아웃라이어가 존재할 때, 이 영향을 최소화
- $x' = \frac{x - \text{median}}{\text{IQR}}$

```python
rb_scaler = RobustScaler()
rb_scaled = rb_scaler.fit_transform(df_wine)
rb_scaled = pd.DataFrame(rb_scaled, columns = df_wine.columns)
rb_scaled
```

```python
round(rb_scaled.describe(), 2)
```

### 3.1.4 MinMaxScaler
- 아웃라이어의 영향을 받음
- 데이터의 기존 분포를 가장 있는 그대로 담아내며 스케일만 변화
- 데이터 범위는 0~1

$
x' = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}
\$

```python
mm_scaler = MinMaxScaler()
mm_scaled = mm_scaler.fit_transform(df_wine)
mm_scaled = pd.DataFrame(mm_scaled, columns = df_wine.columns)
mm_scaled
```

```python
round(mm_scaled.describe(), 2)
```

### 3.15 Modeling

```python
# Split Dataset
X_train, X_test, y_train, y_test = train_test_split(df_wine.drop('class', axis=1), df_wine['class'], test_size=0.2, random_state=100)
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)
```

```python
X_train_scaled = mm_scaler.transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train_scaled, y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred)
```

```python
scores = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled, y_train)
    pred = knn.predict(X_test_scaled)
    acc = accuracy_score(y_test, pred)
    scores.append(acc)
```

```python
sns.lineplot(x=range(1, 21), y=scores);
```

```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/spam.csv'
df_spam = pd.read_csv(file_url)
df_spam.head()
```

```python
df_spam['target'].value_counts()
```

```python
import string
string.punctuation
```

```python
sample_string = df_spam['text'].loc[0]
sample_string
```

```python
new_string = []
for i in sample_string:
    if i not in string.punctuation:
        new_string.append(i)
```

```python
new_string = ''.join(new_string)
new_string
```

```python
def remove_punc(x):
    new_string = []
    for i in x:
        if i not in string.punctuation:
            new_string.append(i)
    new_string = ''.join(new_string)
    return new_string
```

```python
df_spam['text'] = df_spam['text'].apply(remove_punc)
df_spam
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
sample_string = df_spam['text'].loc[0]
sample_string.split()
```

```python
def stop_words(x):
    new_string = []
    for i in x.split():
        if i.lower() in stopwords.words('english'):
            new_string.append(i.lower())
    new_string = ' '.join(new_string)
    return new_string
```

```python
df_spam['text'] = df_spam['text'].apply(stop_words)
df_spam
```

```python
df_spam['target'] = df_spam['target'].map({'spam': 1, 'ham':0})
df_spam
```

```python
X = df_spam['text']
y = df_spam['target']

from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
cv.fit(X)
cv.vocabulary_
```

```python
X = cv.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=100)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
accuracy_score(y_test, pred)
```

```python
print(confusion_matrix(y_test, pred))
```

```python
sns.heatmap(confusion_matrix(y_test, pred), annot=True, fmt= '.0f')
```

## 6.1 중고차 가격 예측

```python
file_url = 'https://media.githubusercontent.com/media/musthave-ML10/data_source/main/car.csv'
df_car = pd.read_csv(file_url)
df_car.head()
```

```python
df_car.info()
```

```python
df_car.isnull().sum()
```

```python
round(df_car.describe(), 2)
```

```python
df_car[['engine', 'engine_unit']] = df_car['engine'].str.split(expand=True)
df_car['engine'] = df_car['engine'].astype('float32')
df_car.drop(columns='engine_unit', axis=1, inplace=True)
```

```python
def isFloat(value):
    try:
        num = float(value)
        return num
    except ValueError:
        return np.NaN
```

```python
df_car[['max_power', 'max_power_unit']] = df_car['max_power'].str.split(expand=True)
df_car['max_power'] = df_car['max_power'].apply(isFloat)
df_car.drop(columns='max_power_unit', axis=1, inplace=True)
```

```python
def mile(x):
    if x['fuel'] == 'Petrol':
        return x['mileage'] / 80.43
    elif x['fuel'] == 'Diesel':
        return x['mileage'] / 73.56
    elif x['fuel'] == 'LPG':
        return x['mileage'] / 40.85
    else: 
        return x['mileage'] / 44.23
```

```python
df_car[['mileage', 'mileage_unit']] = df_car['mileage'].str.split(expand=True)
df_car['mileage'] = df_car['mileage'].astype('float32')
df_car['mileage'] = df_car['mileage'].apply(mile)
df_car.drop(columns='mileage_unit', axis=1, inplace=True)
```
